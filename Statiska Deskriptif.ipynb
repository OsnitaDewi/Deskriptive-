```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('laptop_data.csv')
df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
hitung = df ["Price"]
modus= hitung.mode()
median = hitung.median()
rata = hitung.mean()
maks = hitung.max()
minm = hitung.min()
standardeviasi = hitung.std()
print ("Modusnya adalah : ",modus[0])
print ("Median adalah : ",median)
print ("Rata-ratanya adalah : ",rata)
print ("Meksimal adalah : ",maks)
print ("Minimum adalah : ",minm)
print ("Standar Deviasi : ",standardeviasi)
```

```python
hitung = df["Ram"]
hitung = pd.to_numeric(hitung.str.extract('(\d+)', expand=False), errors='coerce')
hitung = hitung.dropna()
modus = hitung.mode()
median = hitung.median()
rata = hitung.mean()
maks = hitung.max()
minm = hitung.min()
standardeviasi = hitung.std()

print ("Modusnya adalah : ",modus[0])
print ("Median adalah : ",median)
print ("Rata-ratanya adalah : ",rata)
print ("Meksimal adalah : ",maks)
print ("Minimum adalah : ",minm)
print ("Standar Deviasi : ",standardeviasi)
```
